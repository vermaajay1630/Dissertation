```python
import pandas as pd
from lxml import html
from bs4 import BeautifulSoup
import requests
pd.set_option('display.max_columns', None)
url = "https://en.wikipedia.org/wiki/List_of_fiction_works_made_into_feature_films_(0%E2%80%939,_A%E2%80%93C)"
r = requests.get(url)
print(r)
html_doc = r.text
soup = BeautifulSoup(html_doc)
soup.select('.wikitable td+ td a')
data1 = pd.DataFrame(columns=['movieList'])
for i in soup.select('.wikitable td+ td a'):
    data1 = data1.append({'movieList': i}, ignore_index = True)
data1
```

```python
from lxml import html
import json
```

```python
movieId = []
movieName = []
for url in finalUrls:
    r = requests.get(url)
    data = r.text
    fetchedData = json.loads(data)
    resultsData = fetchedData['results']
    for i in resultsData:
        movieId.append(i['id'])
movieId
```

```python
movieName = []
for url in finalUrls:
    r = requests.get(url)
    data = r.text
    fetchedData = json.loads(data)
    resultsData = fetchedData['results']
    for i in resultsData:
        movieName.append(i['original_title'])
movieName
```

```python
movieRelease = []
for url in finalUrls:
    r = requests.get(url)
    data = r.text
    fetchedData = json.loads(data)
    resultsData = fetchedData['results']
    for i in resultsData:
        movieRelease.append(i['release_date'])
movieRelease
```

```python
movieRating = []
for url in finalUrls:
    r = requests.get(url)
    data = r.text
    fetchedData = json.loads(data)
    resultsData = fetchedData['results']
    for i in resultsData:
        movieRating.append(i['vote_average'])
movieRating
```
